```julia
# moments
using Random

totalmoments = 5 # amount of items in momvec
N = 1000000 # try 1m later

function generatemoments(totalmoments, N) # making this in a function because of memory efficiency 
    seed = 12345
    rng = MersenneTwister(seed)

    momvec = zeros(Float64, 0) # the output vector. Will have items pushed onto it.
    momarr = randn(rng, Float64, (totalmoments, N)) # generates a 5x100 matrix of (pseudo)random values to calculate moments

    for i in 1:totalmoments

        moment = 0 # ensure everything is 0 before we start
        
        # the actual algorithm
        for j in 1:N
            moment = moment + momarr[i,j]^i
        end
        moment = moment/N
        
        push!(momvec, moment) # append new value
    end

    return momvec
end

# @time generatemoments(totalmoments, N)

using Plots
gr()

histogram(generatemoments(totalmoments, N))
```

```julia
using Random

steps = 100000
bias = .5

position = 0

seed = 5431
Random.seed!(seed)
# rng = MersenneTwister(seed)

poshist = zeros(Float64, 0)

for i in 1:steps
    randval = rand()
    if randval > bias
        position = position + 1
    else
        position = position - 1
    end

    push!(poshist, position)
    # println(position)
end

print(poshist[steps])

using Plots
gr()
histogram(poshist)

# CONCLUSION: THIS IS NOT A TRNG. THIS IS A PRNG. I am actually graphing the randomness of a function, as opposed to anything else.
# TLDR: My poor entropy :(
```
